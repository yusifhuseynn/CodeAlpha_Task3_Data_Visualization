```python
import pandas as pd

df = pd.read_csv(r"C:\Users\user\airlines_flights_data.csv")
print(df.head())
```

```python
# Number of rows and columns
print(df.shape)

# Data types and missing values
print(df.info())

# Statistical summary (for numerical columns)
print(df.describe())

# View column names
print(df.columns)
```

```python
# Most frequently used airlines
print(df['airline'].value_counts())

# Most common source cities
print(df['source_city'].value_counts())

# Minimum, maximum, and average prices
print(df['price'].min(), df['price'].max(), df['price'].mean())
```

```python
import matplotlib.pyplot as plt

df['airline'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Number of Flights by Airline")
plt.xlabel("Airline")
plt.ylabel("Number of Flights")
plt.show()
```

```python
import seaborn as sns

sns.histplot(df['price'], kde=True, color='orange')
plt.title("Ticket Price Distribution")
plt.show()
```

```python
sns.barplot(x='airline', y='price', data=df, estimator='mean', ci=None)
plt.title("Average Prices by Airline")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='departure_time', y='price', data=df)
plt.title("Gedi vaxtna gr qiymt paylanmas")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='source_city', y='price', hue='destination_city', data=df)
plt.title("Price Comparison by Cities")
plt.show()

```

```python
sns.scatterplot(x='days_left', y='price', data=df)
plt.title("Relationship Between Days Left and Ticket Price")
plt.show()
```
